#**Inicializacion de variables funcion API:**

```python
client_credentials_manager = SpotifyClientCredentials(client_id=cid_spotify, client_secret=secret_spotify)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
genius = lg.Genius(api_key)
```

#**Clean Lyrics:**

```python
def clean_lyrics(lyrics):
  lyrics = re.sub(r"[\[].*?[\]]", "", lyrics)
  lyrics = re.sub(r"[\n]"," ",lyrics)
  return lyrics
```

#**Search Spotify ID Song:**

```python
def get_spotify_song_id(artist,track):
  track_id = sp.search(q='artist:' + artist + ' track:' + track, type='track')
  trackId = track_id['tracks']['items'][0]['id']
  return trackId
```

#**Get Song Popularity:**

```python
def get__song_popularity(artist,track):
  song = sp.search(q='artist:' + artist + ' track:' + track, type='track')
  popularity = song['tracks']['items'][0]['popularity']
  return popularity
```

#**Get Song Duration:**

```python
def get_song_duration(artist,track):
  track_id = sp.search(q='artist:' + artist + ' track:' + track, type='track')
  duration_ms = track_id['tracks']['items'][0]['duration_ms']
  return duration_ms
```

#**BILLBOARD TOP: + Lyrics**

```python
date_list = []
artist_list = []
title_list = []
position_list = []
lyrics_list = []
spotify_ID_list = []
danceability_list = []
energy_list = []
key_list = []
loudness_list = []
speechiness_list = []
acousticness_list = []
instrumentalness_list = []
liveness_list = []
valence_list = []
tempo_list = []
duration_ms_list = []
popularity_list = []
image_url_list = []

weeks_list = []
peakPos_list = []
lastPos_list = []
rank_list = []
isNew_list = []



for year in range(2022,2023):#YEAR
  for month in range(1,5):#MONTH

    date = str(year)+"-"+str(month).zfill(2)+"-01"
    print(date)
    chart = billboard.ChartData('hot-100',date)

    for z in range(10):
        song = chart[z] 
        position = z 

        artist = song.artist
        title = song.title

        weeks = song.weeks
        peakPos = song.peakPos
        lastPos = song.lastPos
        rank = song.rank
        isNew = song.isNew

        try:
          lyrics = genius.search_song(song.title,song.artist).lyrics
          lyrics = clean_lyrics(lyrics)
        except:
          lyrics = "null"

        try:
           image_url = genius.search_song(song.title,song.artist).song_art_image_url
        except:
          image_url = "null"

        date_list.append(date)
        artist_list.append(artist)
        title_list.append(title)
        position_list.append(position)
        lyrics_list.append(lyrics)
        weeks_list.append(weeks) 
        peakPos_list.append(peakPos) 
        lastPos_list.append(lastPos) 
        rank_list.append(rank) 
        isNew_list.append(isNew) 




        try:
          spotify_ID = get_spotify_song_id(artist,title)
        except:
          spotify_ID = "null"
        try:
          duration_ms = get_song_duration(artist,title)
        except:
          duration_ms = "null"


        try:
          popularity = get__song_popularity(artist,title)
        except:
          popularity = "null"
 

        spotify_ID_list.append(spotify_ID)
        duration_ms_list.append(duration_ms)
        popularity_list.append(popularity)
        image_url_list.append(image_url)
        try:
          for x in sp.audio_features(spotify_ID):
              danceability_list.append(x['danceability'])
              energy_list.append(x['energy'])
              key_list.append(x['key'])
              loudness_list.append(x['loudness'])
              speechiness_list.append(x['speechiness'])
              acousticness_list.append(x['acousticness'])
              instrumentalness_list.append(x['instrumentalness'])
              liveness_list.append(x['liveness'])
              valence_list.append(x['valence'])
              tempo_list.append(x['tempo'])
        except:
          danceability_list.append(None)
          energy_list.append(None)
          key_list.append(None)
          loudness_list.append(None)
          speechiness_list.append(None)
          acousticness_list.append(None)
          instrumentalness_list.append(None)
          liveness_list.append(None)
          valence_list.append(None)
          tempo_list.append(None)
      
df = pd.DataFrame({
          'date':date_list,
          'artist':artist_list,
          'title':title_list,
          'position':position_list,
          'lyrics':lyrics_list,
          'spotify id': spotify_ID_list,
          'danceability':danceability_list,
          'energy':energy_list,
          'key':key_list,
          'loudness':loudness_list,
          'speechiness':speechiness_list,
          'acousticness':acousticness_list,
          'instrumentalness':instrumentalness_list,
          'liveness':liveness_list,
          'valence':valence_list,
          'tempo':tempo_list,
          'duration_ms': duration_ms_list, 
          'popularity': popularity_list,
          'image_url':image_url_list,
          'weeks':weeks_list,
          'peakPos':peakPos_list,
          'lastPos':lastPos_list,
          'rank':rank_list,
          'isNew':isNew_list
          
})

```
