#**Get album function**

```python
def get_album(album_id):
    
    album_features_list = ["artist","track_name", "track_id","explicit","is_local","preview_url","track_number","danceability","energy","key","loudness","mode", "speechiness","instrumentalness","liveness","valence","tempo", "duration_ms","time_signature"]
    
    album_df = pd.DataFrame(columns = album_features_list)
    
  
    album = sp.album_tracks("3RQQmkQEvNCY4prGKE6oc5")['items']
    for track in album:
        album_features = {}
        album_features["artist"] =  track['artists'][0]['name']
        album_features["track_name"] = track["name"]
        album_features["track_id"] = track["id"]

        album_features["explicit"] = track["explicit"]
        album_features["is_local"] = track["is_local"]
        album_features["preview_url"] = track["preview_url"]
        album_features["track_number"] = track["track_number"]

        # Audio features
        audio_features = sp.audio_features(album_features["track_id"])[0]
        for feature in album_features_list[8:]:
            album_features[feature] = audio_features[feature]
        
        # Concat audio features df with previous df
        track_df = pd.DataFrame(album_features, index = [0])
        album_df = pd.concat([album_df, track_df], ignore_index = True)
        
        
    return album_df
```
